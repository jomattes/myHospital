```python
resp = requests.get('http://api.us.socrata.com/api/catalog/v1?domains=data.medicare.gov')
data_catalog = pd.DataFrame(columns = ['name'
                              ,'id'
                              ,'download_count'
                              ,'createdAt'
                              ,'data_updated_at'
                              ,'description'
                              ,'columns_field_name'
                             ])
## loop through json response to get each attribute
for i,r in enumerate(resp.json()['results']):
    ## data to become columns in new df    
    data_columns_field_name = r['resource']['columns_field_name']
    
    ## duplicate rows to match col_count
    data_name = np.repeat(r['resource']['name'],len(data_columns_field_name))
    data_id = np.repeat(r['resource']['id'],len(data_columns_field_name))
    data_download_count = np.repeat(r['resource']['download_count'],len(data_columns_field_name))
    data_createdAt = np.repeat(r['resource']['createdAt'],len(data_columns_field_name))
    data_data_updated_at = np.repeat(r['resource']['data_updated_at'],len(data_columns_field_name))
    data_description = np.repeat(r['resource']['description'],len(data_columns_field_name))

    
    ## reformat to list
    list_to_append = list(zip(data_name
                              ,data_id
                              ,data_download_count
                              ,data_createdAt
                              ,data_data_updated_at
                              ,data_description
                              ,data_columns_field_name
                             ))
    ## create df to hold intermediate data
    temp_df = pd.DataFrame(list_to_append, columns = ['name'
                                                  ,'id'
                                                  ,'download_count'
                                                  ,'createdAt'
                                                  ,'data_updated_at'
                                                  ,'description'
                                                  ,'columns_field_name'
                                                 ])
    ## append to new master df
    data_catalog = pd.concat([data_catalog, temp_df])
```
